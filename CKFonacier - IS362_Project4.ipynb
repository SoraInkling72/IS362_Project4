### Adjusting the DataFrame

To simplify things, I will first remove unnecessary columns from the 'fungi' DataFrame, leaving only three, the "Edibility" column, which will be used as the "**target** class"; "Odor"; and a column of my choice, which will be "Cap Shape."

```python
fungi = fungi[['Edibility', 'Odor', 'Cap Shape']]
fungi
```

First the 'target' is the class (series) we want to predict, we would need to drop/remove it from the dataset.

```python
target = 'Edibility' # Class to predict, whether a mushroom is edible or not
traits = fungi[target]

characteristics = fungi.drop(target, axis=1) # This is to drop/remove the class from the
                                             # dataset. Since it's a column, not an index, 
                                             # axis=1 is required for vertical arrays.
```

Since these traits are defined by categories, they cannot be inputted into the scikit classfiers as they are. Apparently this technique, **One Hot Encoding** is a binary feature will convert all of the values for each category into a format that may be fed into machien learning algorithms to improve prediction accuracy.

```python
categories = characteristics.columns # Since all these fungi traits are categorical, the ".columns" method will be used
characteristics = pd.concat([characteristics, pd.get_dummies(characteristics[categories])],
                            axis=1) # This is the code that utilizes "One Hot Encoding," converting values in
                                    # all categories into encoded ones
characteristics.drop(categories, axis=1, inplace=True) # Removes original trait and leaves the encoded values

traits = pd.get_dummies(traits)[1] # This is to encode the target value, in which fungi with
                                   # the value in 'Edibility.' 1 is poisonous and 0 is safe/edible
```

### Testing and Training the Data Models

In this training model, I would have to split the model into training and testing. According to various sources, the **RandomForestClassifier** model is the ideal algorithm to use because I wish for my results to take in the entire dataset, which is large, and have a high accuracy rate. Hence, Random Forest would be the best training model to use in comparison to KNeighbors, which may have a likelihood of producing imprecise results.

The cell below splits the datset into training and testing.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(characteristics, traits, random_state = 0)
```

The following two code cells make use of both the testing and training model

#### Testing Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_test, y_test)

test_model = sorted(zip(model.feature_importances_,X_test.columns))
test_model.sort(reverse=True)
test_model
```

#### Training Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

train_model = sorted(zip(model.feature_importances_,X_train.columns))
train_model.sort(reverse=True)
train_model
```
